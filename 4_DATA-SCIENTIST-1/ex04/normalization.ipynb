## Data Normalization: Making Your Data Speak the Same Language

Data normalization is a crucial preprocessing step in machine learning that ensures all features contribute equally to model training. When features have vastly different scales - like age (0-100) versus income (0-100,000) - algorithms can become biased toward larger-scale features.

### Why Normalize Data?

Raw data often comes in different units and scales. Without normalization:
- Features with larger ranges dominate the learning process
- Distance-based algorithms (like KNN, SVM) perform poorly
- Gradient descent converges slowly or fails to converge

### Scikit-Learn's Normalization Arsenal

**StandardScaler** (Z-score normalization) transforms data to have mean=0 and std=1:
- Best for normally distributed data
- Formula: (x - mean) / standard_deviation

**MinMaxScaler** scales features to a fixed range [0,1]:
- Preserves relationships between values
- Sensitive to outliers

**RobustScaler** uses median and interquartile range:
- Less sensitive to outliers
- Good for data with extreme values

### Implementation Pattern

The typical workflow involves:
1. Fit the scaler on training data only
2. Transform both training and test sets using the same scaler
3. This prevents data leakage and ensures consistent scaling

### Real Impact

Notice how our knight classification data becomes more balanced after scaling. Features like 'Empowered' and 'Stims' now contribute equally to the model, revealing patterns that were previously hidden by scale differences.

Normalization isn't just a technical step - it's about giving every feature a fair voice in your model's decision-making process.

## Normalization vs Standardization: Understanding the Scaling Spectrum

When preparing data for machine learning, **normalization** and **standardization** are often used interchangeably, but they represent distinct approaches to feature scaling with different mathematical foundations and use cases.

### The Core Difference

**Standardization (Z-score normalization)** transforms data to have a mean of 0 and standard deviation of 1:
- Formula: `z = (x - μ) / σ`
- Results in data centered around zero
- Maintains the original distribution shape
- Not bounded to a specific range

**Normalization (Min-Max scaling)** rescales data to a fixed range, typically [0,1]:
- Formula: `x_norm = (x - min) / (max - min)`
- Results in data bounded between 0 and 1
- Changes the distribution shape
- All values fall within the specified range

### When to Use Each Method

**Choose Standardization when:**
- Your data follows a normal distribution
- You're using algorithms sensitive to feature scale (SVM, Neural Networks, PCA)
- Outliers shouldn't be compressed into a narrow range
- You want to preserve the relative distances between data points

**Choose Normalization when:**
- You need bounded outputs (neural network activations)
- Features have known min/max boundaries
- You're comparing features with different units
- You want all features to contribute equally regardless of their original scale

### The Feature Scaling Family Tree

Both normalization and standardization belong to the broader category of **feature scaling** techniques:

1. **StandardScaler** - Zero mean, unit variance
2. **MinMaxScaler** - Bounded range scaling
3. **RobustScaler** - Median-based, outlier-resistant
4. **MaxAbsScaler** - Scale by maximum absolute value
5. **PowerTransformer** - Make data more Gaussian-like

### Real-World Impact

Consider our knight dataset: without scaling, features like 'Power' (ranging 100-2000) would dominate 'Agility' (ranging 0.05-0.3). Standardization ensures each Jedi's 'Agility' has equal voice with their 'Power' in determining their classification.

The choice between normalization and standardization isn't just technical—it's about how you want your algorithm to perceive and weight your features. Standardization says "treat all features democratically," while normalization says "compress everything to the same playing field."

### Best Practices

- **Always fit on training data only** to prevent data leakage
- **Apply the same transformation to test data** using the training scaler
- **Consider your algorithm's assumptions** when choosing scaling methods
- **Visualize before and after** to understand the transformation's impact

Remember: scaling doesn't change relationships between data points—it changes how algorithms perceive and process those relationships.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
X_train = df_train.drop('knight', axis=1)
X_test = df_test

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

df_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
df_train_scaled['knight'] = df_train['knight'].values  # Add back the target column

df_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)

df_train_scaled = df_train_scaled.round(2)
df_test_scaled = df_test_scaled.round(2)
```

```python
df_train_scaled.head(15)
```

```python
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(12, 10))
sns.scatterplot(data=df_train_scaled, x='Empowered', y='Stims', hue='knight', ax=ax[0,0], palette=['blue', 'red'], alpha=0.3)
sns.scatterplot(data=df_test_scaled, x='Empowered', y='Stims', ax=ax[1,0])
sns.scatterplot(data=df_train_scaled, x='Push', y='Deflection', hue='knight', ax=ax[0,1], palette=['blue', 'red'], alpha=0.42)
sns.scatterplot(data=df_test_scaled, x='Push', y='Deflection', ax=ax[1,1])
plt.show()
```
