## Data Standardization: Making Your Data Speak the Same Language
> ..."putting all variables on a level playing field"

#### Data standardization
- crucial preprocessing step in dataset transformation. 
- This means all features have the same:


    1. Scale
    2. Distribution.


### Why Standardize?

When working with datasets like our Knight data, you'll notice that features have vastly different ranges:
- `Sensitivity` ranges from ~7 to ~20
- `Power` ranges from ~100 to ~2400
- `Agility` ranges from ~0.05 to ~0.25

These different scales can cause problems in machine learning algorithms, where features with larger values might dominate those with smaller values, leading to biased results.

### Common Standardization Methods

**Z-Score Normalization (Standard Scaler):**
- Transforms data to have mean = 0 and standard deviation = 1
- Formula: `(x - mean) / standard_deviation`
- Best for normally distributed data

**Min-Max Scaling:**
- Scales data to a fixed range, typically [0,1]
- Formula: `(x - min) / (max - min)`
- Preserves relationships between values

**Robust Scaling:**
- Uses median and interquartile range instead of mean and standard deviation
- Less sensitive to outliers

### When to Standardize

Standardization is essential for:
- Distance-based algorithms (KNN, K-means, SVM)
- Neural networks
- Principal Component Analysis (PCA)
- Gradient descent optimization

Your Knight dataset is a perfect candidate for standardization due to its diverse feature scales and the likely need for machine learning analysis.

### Practical Implementation with Scikit-Learn

Now let's apply standardization to our Knight datasets using Python's scikit-learn library:

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

```

## Comparing Data Standardization Methods

### StandardScaler (Z-Score Normalization)
**What it does:** Transforms data to have a mean of 0 and standard deviation of 1
- **Formula:** `(x - mean) / standard_deviation`
- **Best for:** Normally distributed data
- **Use case:** Most machine learning algorithms, especially those sensitive to feature scales
- **Result:** Data centered around 0, with most values between -3 and +3

### MinMaxScaler
**What it does:** Scales data to a fixed range, typically [0,1]
- **Formula:** `(x - min) / (max - min)`
- **Best for:** When you need bounded values within a specific range
- **Use case:** Neural networks, image processing, when you want to preserve exact relationships
- **Result:** All values squeezed between 0 and 1 (or your chosen range)

### RobustScaler
**What it does:** Uses median and interquartile range instead of mean and standard deviation
- **Formula:** `(x - median) / IQR`
- **Best for:** Data with outliers
- **Use case:** When your dataset has extreme values that would distort StandardScaler or MinMaxScaler
- **Result:** Less sensitive to outliers, data centered around 0

### Key Differences Summary:
- **StandardScaler:** Most common, assumes normal distribution
- **MinMaxScaler:** Preserves exact zero values, bounded output
- **RobustScaler:** Outlier-resistant, uses robust statistics

**For your Knight dataset:** StandardScaler is typically the best choice since it works well with most machine learning algorithms and your data appears to have reasonable distributions without extreme outliers.
