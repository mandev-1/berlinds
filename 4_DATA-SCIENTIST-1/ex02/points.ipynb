## Train Dataset Analysis

The training dataset reveals distinct correlation patterns among the features:

### Strong Positive Correlations with Target
- **Empowered**: 0.794 - Highest correlation, indicating strong predictive power
- **Prescience**: 0.790 - Nearly equivalent correlation strength
- **Stims**: 0.787 - Strong correlation suggesting importance in classification
- **Recovery**: 0.778 - Significant positive correlation

### Weak Positive Correlations
- **Repulse**: 0.324 - Moderate correlation
- **Friendship**: 0.237 - Weak positive correlation
- **Mass**: 0.113 - Minimal positive correlation
- **Midi-chlorien**: 0.008 - Nearly zero correlation

### Negative Correlations (Minimal)
- **Push**: -0.019 - Slight negative correlation
- **Deflection**: -0.026 - Minimal negative correlation
- **Survival**: -0.043 - Weakest negative correlation

The analysis suggests that Force-related abilities (Empowered, Prescience) and enhancement capabilities (Stims, Recovery) are the strongest predictors for knight classification, while physical attributes show weaker correlations.

```python
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(12, 10))
sns.scatterplot(data=df_train, x='Empowered', y='Stims', hue='knight', ax=ax[0,0], palette=['blue', 'red'], alpha=0.3)
sns.scatterplot(data=df_test, x='Empowered', y='Stims', ax=ax[1,0])
sns.scatterplot(data=df_train, x='Push', y='Deflection', hue='knight', ax=ax[0,1], palette=['blue', 'red'], alpha=0.42)
sns.scatterplot(data=df_test, x='Push', y='Deflection', ax=ax[1,1])
plt.show()
```
