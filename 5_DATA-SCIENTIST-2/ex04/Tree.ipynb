```python
## QUICKSTART CHUNK (from ex03 of this assignment)

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler

df_training = pd.read_csv('data/Training_knight.csv')
df_validati = pd.read_csv('data/Validation_knight.csv')
```

```python
# Knight Classification for Jupyter Notebook
# Use this after loading your data with:
# df_training = pd.read_csv('data/Training_knight.csv')
# df_validati = pd.read_csv('data/Validation_knight.csv')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

def prepare_data(df_train, df_test):
    """Prepare the data for training and testing"""
    print(f"Training data shape: {df_train.shape}")
    print(f"Validation data shape: {df_test.shape}")
    
    # Prepare training data
    X_train = df_train.drop('knight', axis=1)
    y_train = df_train['knight']
    
    # Prepare test data
    if 'knight' in df_test.columns:
        X_test = df_test.drop('knight', axis=1)
        y_test = df_test['knight']
    else:
        X_test = df_test
        y_test = None
    
    # Handle missing values
    X_train = X_train.fillna(X_train.mean())
    X_test = X_test.fillna(X_train.mean())
    
    print(f"Features: {X_train.shape[1]}")
    print(f"Classes: {sorted(y_train.unique())}")
    print(f"Class distribution:\n{y_train.value_counts()}")
    
    return X_train, y_train, X_test, y_test

def train_random_forest(X_train, y_train):
    """Train Random Forest model"""
    rf_model = RandomForestClassifier(
        n_estimators=200,
        max_depth=20,
        min_samples_split=2,
        min_samples_leaf=1,
        random_state=42,
        n_jobs=-1
    )
    
    rf_model.fit(X_train, y_train)
    
    # Evaluate with cross-validation
    cv_scores = cross_val_score(rf_model, X_train, y_train, cv=5, scoring='f1_macro')
    print(f"Random Forest CV F1-score: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
    
    return rf_model, cv_scores.mean()

def train_decision_tree(X_train, y_train):
    """Train Decision Tree model"""
    dt_model = DecisionTreeClassifier(
        max_depth=15,
        min_samples_split=5,
        min_samples_leaf=2,
        criterion='gini',
        random_state=42
    )
    
    dt_model.fit(X_train, y_train)
    
    # Evaluate with cross-validation
    cv_scores = cross_val_score(dt_model, X_train, y_train, cv=5, scoring='f1_macro')
    print(f"Decision Tree CV F1-score: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
    
    return dt_model, cv_scores.mean()

def visualize_tree(model, feature_names, class_names, model_name):
    """Visualize the decision tree"""
    plt.figure(figsize=(25, 15))
    
    if hasattr(model, 'estimators_'):  # Random Forest
        tree_to_plot = model.estimators_[0]
        title = f"Decision Tree from {model_name} (First Tree)"
    else:  # Decision Tree
        tree_to_plot = model
        title = f"{model_name}"
    
    plot_tree(tree_to_plot,
              feature_names=feature_names,
              class_names=class_names,
              filled=True,
              rounded=True,
              fontsize=10,
              max_depth=4)
    
    plt.title(title, fontsize=20, fontweight='bold')
    plt.tight_layout()
    plt.show()

def plot_feature_importance(model, feature_names, model_name):
    """Plot feature importance"""
    if hasattr(model, 'feature_importances_'):
        importance_df = pd.DataFrame({
            'feature': feature_names,
            'importance': model.feature_importances_
        }).sort_values('importance', ascending=False)
        
        print(f"\nTop 10 Most Important Features ({model_name}):")
        print(importance_df.head(10))
        
        # Plot feature importance
        plt.figure(figsize=(12, 8))
        top_features = importance_df.head(15)
        plt.barh(range(len(top_features)), top_features['importance'])
        plt.yticks(range(len(top_features)), top_features['feature'])
        plt.xlabel('Feature Importance')
        plt.title(f'Top 15 Feature Importances - {model_name}')
        plt.gca().invert_yaxis()
        plt.tight_layout()
        plt.show()
        
        return importance_df
    return None

def plot_confusion_matrix(y_true, y_pred, class_names, model_name):
    """Plot confusion matrix"""
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
               xticklabels=class_names,
               yticklabels=class_names)
    plt.title(f'Confusion Matrix - {model_name}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.show()

def evaluate_model(model, X_test, y_test, model_name):
    """Evaluate model performance"""
    predictions = model.predict(X_test)
    
    if y_test is not None:
        f1 = f1_score(y_test, predictions, average='macro')
        print(f"\n{model_name} Test Results:")
        print(f"F1-score: {f1:.4f}")
        print("\nClassification Report:")
        print(classification_report(y_test, predictions))
        
        # Plot confusion matrix
        class_names = sorted(y_test.unique())
        plot_confusion_matrix(y_test, predictions, class_names, model_name)
        
        return predictions, f1
    
    return predictions, None

# Main execution
print("Knight Classification System")
print("=" * 50)

# Prepare data using your existing dataframes
X_train, y_train, X_test, y_test = prepare_data(df_training, df_validati)

# Train both models
print("\nTraining models...")
rf_model, rf_cv_score = train_random_forest(X_train, y_train)
dt_model, dt_cv_score = train_decision_tree(X_train, y_train)

# Evaluate on training data
rf_train_pred = rf_model.predict(X_train)
dt_train_pred = dt_model.predict(X_train)

rf_train_f1 = f1_score(y_train, rf_train_pred, average='macro')
dt_train_f1 = f1_score(y_train, dt_train_pred, average='macro')

print(f"\nTraining F1-scores:")
print(f"Random Forest: {rf_train_f1:.4f}")
print(f"Decision Tree: {dt_train_f1:.4f}")

# Choose best model based on cross-validation scores
if rf_cv_score >= dt_cv_score:
    best_model = rf_model
    best_model_name = "Random Forest"
    best_f1 = rf_train_f1
else:
    best_model = dt_model
    best_model_name = "Decision Tree"
    best_f1 = dt_train_f1

print(f"\nBest model: {best_model_name} (Training F1: {best_f1:.4f})")

# Check if meets requirement
if best_f1 >= 0.90:
    print("✅ Model meets the minimum 90% F1-score requirement!")
else:
    print("⚠️  Model does not meet the minimum 90% F1-score requirement.")

# Make predictions and evaluate
test_predictions, test_f1 = evaluate_model(best_model, X_test, y_test, best_model_name)

# Visualize the tree
print("\nGenerating visualizations...")
feature_names = X_train.columns.tolist()
class_names = sorted(y_train.unique())

visualize_tree(best_model, feature_names, class_names, best_model_name)

# Plot feature importance
importance_df = plot_feature_importance(best_model, feature_names, best_model_name)

# Save predictions to Tree.txt
with open('Tree.txt', 'w') as f:
    for pred in test_predictions:
        f.write(f"{pred}\n")

print(f"\nPredictions saved to Tree.txt")
print(f"Total predictions: {len(test_predictions)}")

# Display first few predictions
print(f"\nFirst 10 predictions:")
for i, pred in enumerate(test_predictions[:10]):
    print(f"{i+1}: {pred}")

print("\n" + "="*50)
print("SUMMARY:")
print(f"Best Model: {best_model_name}")
print(f"Training F1-score: {best_f1:.4f}")
if test_f1:
    print(f"Test F1-score: {test_f1:.4f}")
print(f"Requirements met: {'✅ Yes' if best_f1 >= 0.90 else '⚠️ No'}")
print("="*50)
```
